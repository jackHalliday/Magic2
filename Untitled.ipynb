```python
import pickle
import gzip
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import numpy as np
%matplotlib notebook
```

```python
%matplotlib notebook
```

```python
with gzip.open("dump_for_paper.pkl", 'rb') as f:
    labels, points, all_t, flat_t = pickle.load(f)
```

```python
ax = plt.figure().subplots()
cm = LinearSegmentedColormap.from_list("custom", np.array([(238,46,47),(64,86,165)])/255)
ax.imshow(np.ma.masked_where(labels == -1024, labels), cmap=cm, clim=(71,72), interpolation="nearest")
width = 2
ax.triplot(points[:, 1], points[:, 0], all_t, lw=width)
ax.triplot(points[:, 1], points[:, 0], flat_t, lw=width)
ax.plot
xlim = np.array((1490,1520))+.5
ax.set_xlim(xlim)
ax.set_xticks(np.arange(*xlim))
ylim = np.array((1090,1075))+.5
ax.set_ylim(ylim)
ax.set_yticks(np.arange(*ylim,-1))
plt.grid(alpha=0.5)
plt.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
plt.tight_layout()
plt.savefig("pixels.pdf")
```
